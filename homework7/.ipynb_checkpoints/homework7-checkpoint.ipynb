```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def make_sine_wave(x, A, w):
    return A * np.sin(w * x)

x = np.linspace(0, 2 * np.pi, 1000)

amplitude = [0.5, 1, 1.5, 2, 2.5]
frequency = [1, 2, 3, 4, 5]
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(25, 5))

for A in amplitude:
    y = make_sine_wave(x, A, 1)
    ax1.plot(x, y, label=f'A={A}')

for w in frequency:
    y = make_sine_wave(x, 1, w)
    ax2.plot(x, y, label=f'w={w}')

for A, w in zip(amplitude, frequency):
    y = make_sine_wave(x, A, w)
    ax3.plot(x, y, label=f'A={A}, w={w}')

ax1.set_title('Amplitude (w=1)', fontsize=20)
ax1.set_xlabel('x', fontsize=12)
ax1.set_ylabel('y', fontsize=12)
ax1.legend(fontsize=10)
ax1.grid(True, alpha=0.3)

ax2.set_title('Frequency (A=1)', fontsize=20)
ax2.set_xlabel('x', fontsize=12)
ax2.set_ylabel('y', fontsize=12)
ax2.legend(fontsize=10)
ax2.grid(True, alpha=0.3)

ax3.set_title('Both Amplitude and Frequency', fontsize=20)
ax3.set_xlabel('x', fontsize=12)
ax3.set_ylabel('y', fontsize=12)
ax3.legend(fontsize=10)
ax3.grid(True, alpha=0.3)

plt.show()
```
