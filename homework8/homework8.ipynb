```python
#4
print(states.shape) #52 rows, 4 columns

#5
print(states.head())
print(states.tail())
"""
The columns represent the shortened 2-letter state name, latitude and longitude, and the full state name
The first state is Alaska
The last state is Wyoming
There are 51 entries in total
Puerto Rico is listed as one of the states even though officially it's a territory
"""
```

```python
#6
print(states["latitude"].median())
print(states["longitude"].median())
#latitude = 39.44
#longitude = -89.09

#7
median_lat = states["latitude"].median()
states["distance_to_median_lat"] = abs(states["latitude"] - median_lat)

#8
median_lon = states["longitude"].median()
states["distance_to_median_lon"] = abs(states["longitude"] - median_lon)

#9
print(states.shape) #52 rows, 6 columns

#10
median_lat = states["latitude"].median()
median_lon = states["longitude"].median()
states["lat_distance"] = abs(states["latitude"] - median_lat)
states["lon_distance"] = abs(states["longitude"] - median_lon)

closest_lat_idx = states["lat_distance"].idxmin()
closest_lat_state = states.loc[closest_lat_idx, "state"]
closest_lat_dist = states.loc[closest_lat_idx, "lat_distance"]
print(f"state closest to median latitude: {closest_lat_state} by {closest_lat_dist:.2f}")

farthest_lat_idx = states["lat_distance"].idxmax()
farthest_lat_state = states.loc[farthest_lat_idx, "state"]
farthest_lat_dist = states.loc[farthest_lat_idx, "lat_distance"]
print(f"state farthest from median latitude {farthest_lat_state} by {farthest_lat_dist:.2f}")

closest_lon_idx = states["lon_distance"].idxmin()
closest_lon_state = states.loc[closest_lon_idx, "state"]
closest_lon_dist = states.loc[closest_lon_idx, "lon_distance"]
print(f"state closest to median longitude {closest_lon_state} by {closest_lon_dist:.2f}")

farthest_lon_idx = states["lon_distance"].idxmax()
farthest_lon_state = states.loc[farthest_lon_idx, "state"]
farthest_lon_dist = states.loc[farthest_lon_idx, "lon_distance"]
print(f"state farthest from median longitude {farthest_lon_state} by {farthest_lon_dist:.2f}")

"""
state closest to median latitude: CO by 0.11
state farthest from median latitude: AK by 24.15
state closest to median longitude: IL by 0.31
state farthest from median longitude: HI by 66.57
"""
```
