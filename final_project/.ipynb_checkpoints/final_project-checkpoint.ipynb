```python
#overview
"""
This code will have a lot of technical details so I will provide a brief overview of what I am trying to do here:
1. Import all the necessary libraries
2. Set the parameters and initialize the dataframe
3. Calculate the nuclear timescale (M/L) as an order-of-magnitude estimate for comparison with the final integrated lifetime
4. Use the Lane-Emden polytrope equation to define a function for pressure and temperature at a given radius R
5. Define the equation used to determine the reaction rate for the proton-proton chain
6. Integrate over the radius of the star using the pressure and temperature functions from (4) as inputs to the equation in (5)
and multiply by the energy per reaction to get a luminosity
7. Multiply luminosity by the "bin width", and calculate how much total energy was used over this timespan
8. Determine the change in hydrogen mass fraction needed to produce said energy
9. Update the data table with the change in hydrogen mass fraction
10. Repeat steps 4-9 in a recursive algorithm until the hydrogen reaches 0
11. Multiply the length of the data frame by the bin width to get an expected lifespan
12. Compare the result to the prediction made in (3) to get an error value
13. A visualization of the simulation will be available after the algorithm terminates
"""
```

```python
#importing libraries
import pandas as pd
import matplotlib.pyplot as plt
import scipy as si
import numpy as np
import matplotlib.patches as patches
from matplotlib.widgets import Slider
from scipy.integrate import solve_ivp, simpson
from astropy import constants as const
from astropy import units as u
from ipywidgets import interact, IntSlider
import matplotlib.patches as patches
```

```python
#Lane-Emden Equation Class - First component of solving for energy

"""
Model approach: Hydrostatic equilibrium (polytrope)+ energy (nuclear timescale)
To maintain hydrostatic equilibrium, the star must have a certain amount of pressure counteracting gravity at a given radius
We can calculate how much energy is released from each of the main reaction types, and how many of those reactions are occuring

defining functions to update the data frame
lane-emden equations with polytrope index n = 1.5 (appropriate for red dwarf stars)
use polytropic approach for simplifications; need definite integrals
use Runge-Kutta to solve the differential equation

IMPORTANT: Credit to Michael Zingale, whose notebook I referenced for the derivation and code at this link: https://zingale.github.io/stars/notebooks/polytrope/lane-emden.html
"""
class LaneEmdenSolver: #Solves the Lane-Emden equation for a polytrope of index n
    
    def __init__(self, n=1.5): #For red dwarf stars, we can just set this to one value.
        self.n = n
        self.xi = None
        self.theta = None
        self.dtheta_dxi = None
        self.xi_surface = None 
        
    def solve(self, xi_max=10):
        y0 = [1.0, 0.0]
        
        def surface_event(xi, y):
            return y[0] 
        surface_event.terminal = True
        surface_event.direction = -1
        
        sol = solve_ivp( 
            fun=self._lane_emden_derivatives,
            t_span=[1e-6, xi_max],
            y0=y0,
            method='RK45',
            events=surface_event,
            dense_output=True,
            max_step=0.1
        )
        
        self.xi = sol.t
        self.theta = sol.y[0]
        self.dtheta_dxi = sol.y[1]
        positive_mask = self.theta > 0
        self.xi = self.xi[positive_mask]
        self.theta = self.theta[positive_mask]
        self.dtheta_dxi = self.dtheta_dxi[positive_mask]
        self.xi_surface = self.xi[-1]
        
        return self.xi, self.theta, self.dtheta_dxi
    
    def _lane_emden_derivatives(self, xi, y):  #we use scipy to create the derivatives for the equation
        theta, dtheta = y
        theta = max(theta, 0.0)
        if xi < 1e-10:
            d2theta = -theta**self.n / 3.0
        else:
            d2theta = -2.0*dtheta/xi - theta**self.n
        return [dtheta, d2theta]


def get_stellar_structure(M_solar, X, polytrope_solution):
    """
    Convert Lane-Emden solution to physical stellar structure.
    -M_solar is stellar mass in solar masses
    -X is hydrogen mass fraction
    -polytrope_solution is solved Lane-Emden equation
    Returns dictionary with keys 'R', 'rho_c', 'T_c', 'r', 'rho', 'T'
    """
    #converting to mass in g
    M = M_solar * M_sun
    
    #Mean molecular weight for fully ionized gas
    Y = 1.0 - X
    mu = 1.0 / (2.0*X + 0.75*Y)
    
    #Mass-radius relation for red dwarfs
    R = (M_solar**0.8) * R_sun
    
    #scale factor, so this works for stars of all radii
    alpha = R / polytrope_solution.xi_surface
    
    #density
    integrand = polytrope_solution.xi**2 * polytrope_solution.theta**polytrope_solution.n
    integral = simpson(integrand, polytrope_solution.xi)
    rho_c = M / (4.0 * np.pi * alpha**3 * integral)
    
    #temperature
    T_virial = G * M * mu * m_p / (k_B * R)
    T_c = 0.6 * T_virial  # Correction factor for n=1.5
    
    #now we can put it all together, and assign these variables to the equations from the L-E solver.
    r = alpha * polytrope_solution.xi
    rho = rho_c * polytrope_solution.theta**polytrope_solution.n
    T = T_c * polytrope_solution.theta

    #return everything as a dictionary so it's easy to retrieve later on!
    return {
        'R': R,
        'rho_c': rho_c,
        'T_c': T_c,
        'r': r,
        'rho': rho,
        'T': T
    }
```

```python
#The energy generated by the proton-proton chain
"""
This equation gives us the relationship between the rate of energy generation in the P-P chain and the variables that are output from
the L-E equations (Density, Temperature), and we can treat mass fraction of hydrogen as a constant over the bin width.

This exact equation was difficult to find; i went through sources like these:
https://www.as.utexas.edu/astronomy/education/spring16/lambert/secure/Lecture14.pdf
http://user.astro.wisc.edu/~townsend/resource/teaching/astro-310-F19/handouts/handout-21.pdf

But there were different values given for the power of the relationship between epsilon and the other 3 variables, so i am not sure
of the accuracy of this
"""
def epsilon_pp(rho, T, X):
    epsilon_0 = 2.4e-6  #UNITS ARE IMPORTANT: these are in erg/g/s (CGS)
    T6 = T / 1e6       #Temperature in units of 10^6 K
    
    return epsilon_0 * rho * X**2 * T6**4
```

```python
#part here to initialize the data frame
"""
parameters
-mass of the star (initial_mass)
-time since simulation started (time_elapsed)
-mass fraction of hydrogen (X)
-mass fraction of helium (Y)
-metallicity (mass fraction of other elements) (Z)
-bin width (bin_width)

assumptions
-Constant mass, luminosity, and radius
-Only red dwarves (100% convective, only P-P chain, only hydrogen fusion)

Users can set the metallicity and mass at the start

Need to set some parameters of the simulation here as well.
"""
bin_width = 1e9 #discrete period over which the simulation runs, in years. For long lived red dwarfs, this should be a large number!
x_final = 0.03 #since this is just an approximation, the mass fraction never exactly reaches zero, so I have to truncate it at a finite value.
max_runs = 1e7 #prevents a runtime error

time = 0
X = 0.7
Y = 0.2
Z = 0.1
initial_mass = 0.5 #a value between 0.1 and 0.5 solar massess, the mass range for fully convective red dwarfs

#some physical constants from astropy. epsilon equation we will be using is in cgs, so i converted everything to cgs.
M_sun = const.M_sun.cgs.value      #g
R_sun = const.R_sun.cgs.value      #cm
L_sun = const.L_sun.cgs.value      #erg/s
G = const.G.cgs.value              #cm^3/g/s^2
k_B = const.k_B.cgs.value          #erg/K
m_p = const.m_p.cgs.value          #g
c = const.c.cgs.value              #cm/s

columns = ['time', 'X', 'Y', 'Z', 'mass', 'T_core', 'rho_core', 'L', 'R'] #The last 4 variables will be needed for the algorithm and are dependent
initial_data = {
    'time': [time],
    'X': [X],
    'Y': [Y],
    'Z': [Z],
    'mass': [initial_mass],
    'T_core': [np.nan],    #These are blank for now and will be initialized when the algorithm starts running.
    'rho_core': [np.nan],  
    'L': [np.nan],         
    'R': [np.nan]          
}
main_df = pd.DataFrame(initial_data)
```

```python
#the recursive algorithm functions
"""
This part is where we actually execute the algorithm. We need to update the elemental composition and luminosity as well. There are
3 functions: two which update the equations of state, and the final one to update the main data frame.
Note: Everything is in cgs.
"""
def update_composition(L_solar, M_solar, X, Y, dt_years):
    """
    Update hydrogen and helium fractions based on energy radiated
    Uses mass-energy conversion E=mc^2
    For H to He fusion, the efficiency eta is around 0.007
    """
    #energy radiated over timestep
    L_erg_s = L_solar * L_sun
    dt_sec = dt_years * 3.156e7
    E_radiated = L_erg_s * dt_sec 
    
    eta = 0.007
    
    #hydrogen mass consumed
    M_H_consumed = E_radiated / (eta * c**2)
    
    #change in mass fraction
    M_star = M_solar * M_sun
    dX = -M_H_consumed / M_star
    
    X_new = X + dX
    Y_new = Y - dX
    
    #This part prevents the code from erroring if X goes negative
    if X_new < 0:
        X_new = 0.0
        Y_new = Y + X
    
    return X_new, Y_new


def calculate_luminosity(structure, X, M_solar):
    """
    Luminosity accounting for both X decrease and T increase; uses empirical M-L relation with composition and temperature corrections
    """
    X_ZAMS = 0.70
    X_current = X
    
    #Base luminosity from M-L relation, which goes as M^3.5
    L_ZAMS = M_solar**3.5
    
    #Mean molecular weight changes with composition, which I realized later on impacts the results of the simulation greatly.
    mu_ZAMS = 1.0 / (2.0 * X_ZAMS + 0.75 * (1.0 - X_ZAMS))
    mu_current = 1.0 / (2.0 * X_current + 0.75 * (1.0 - X_current))
    
    temperature_factor = (mu_current / mu_ZAMS)**4
    composition_factor = (X_current / X_ZAMS)**2
    
    L = L_ZAMS * composition_factor * temperature_factor
    
    return L


def evolve_red_dwarf(M_solar, X_initial, X_final, dt_years = bin_width, max_steps = max_runs, Y_initial = Y, Z_initial = Z):
    print(f"Evolving {M_solar} M_sun red dwarf")
    
    solver = LaneEmdenSolver(n=1.5)
    solver.solve()
    
    X = X_initial
    Y = Y_initial
    Z = Z_initial
    time = 0.0
    
    #Determine stellar structure at t=0
    structure = get_stellar_structure(M_solar, X, solver)
    L = calculate_luminosity(structure, X, M_solar)
    
    df_data = {
        'time': [0.0],
        'X': [X],
        'Y': [Y],
        'Z': [Z],
        'mass': [M_solar],
        'T_core': [structure['T_c']],
        'rho_core': [structure['rho_c']],
        'L': [L],
        'R': [structure['R'] / R_sun]
    }
    evolution_df = pd.DataFrame(df_data)
    
    step = 0
    
    while X > X_final and step < max_steps:
        structure = get_stellar_structure(M_solar, X, solver)
        L = calculate_luminosity(structure, X, M_solar)
        X_new, Y_new = update_composition(L, M_solar, X, Y, dt_years)
        time += dt_years
        
        #Store data
        new_row = {
            'time': time / 1e9,
            'X': X_new,
            'Y': Y_new,
            'Z': Z,
            'mass': M_solar,
            'T_core': structure['T_c'],
            'rho_core': structure['rho_c'],
            'L': L,
            'R': structure['R'] / R_sun
        }
        evolution_df = pd.concat([evolution_df, pd.DataFrame([new_row])], ignore_index=True)
        
        #Update for next iteration
        X = X_new
        Y = Y_new
        step += 1
        
        if step % 100 == 0:
            print(f"Step {step:5d}: t={time/1e9:6.2f} Gyr, X={X:.4f}")
    print(f"Total time: {time/1e9:.1f} Gyr")
    print(f"Error ratio: {(time/1e9)/control_timescale:.1f}")
    return evolution_df
```

```python
#Control output 
"""
This cell runs predictions using the sun's known lifespan, the nuclear timescale, and the mass-luminosity relationship of stars to determine
the expected lifespan of the star.

The nuclear timescale uses the fusion rate within a star to relate luminosity to mass. The star "dies" when the fuel has been exhausted,
and thus scales as M/L, or how much the star radiates vs. how much fuel it can burn.
If the luminosity of the star stays constant throughout its lifetime, then luminosity scales roughly as L ~ M^3.5.
(Important note on the above point: the luminosity of the star will drastically increase during the giant phases, so the nuclear timescale
values are typically said to give the main-sequence lifespan of a star, though for the purposes of this project, that will not matter so much,
since red dwarfs never enter the giant phase of their life.)
We can also assume the fuel the star has scales LOCALLY (<0.5 M_sun) proportionally to M. An important nuance here is that the sun, the benchmark star,
does NOT have a fully convective interior, and thus only fuses about 10% of its available hydrogen, whereas a red dwarf is fully convective
and will use all of its available fuel. We need to correct by multiplying the nuclear timescale by 10.

Putting these components together, we get that lifespan goes as M/M^3.5 ~ M^-2.5.

The sun has an expected lifespan of 10^10yr (10Gyr). We can use a simple ratio that takes the mass of the input star and returns its expected lifespan.
"""
def nuclear_timescale(mass):
    lifetime_years = (1 / (mass ** 2.5)) * 1e10 * 10
    lifetime_gyr = lifetime_years / 1e9
    
    return round(lifetime_gyr, 2)

control_timescale = nuclear_timescale(initial_mass) #we will compare this to the actual value we get later!
print(f"Nuclear timescale for {initial_mass} M_sun: {control_timescale} Gyr")
```

```python
#Running the algorithm
result_df = evolve_red_dwarf(
    M_solar=initial_mass, 
    X_initial=X,          
    X_final=0.03,
    dt_years=bin_width,          
    max_steps=100000
)

main_df = result_df
```

```python
#Now that the simulation has run, we can use graphs to represent the data.
"""
There are two graphs that accompany the simulation:
(1) A bar chart, giving an overview of the internal composition for the entire lifespan of the star
(2) An interactive cross-section diagram, showing a hypothetical cutout of the star at any given time, which can be changed via a slider
"""
def plot_composition_evolution(df, M_solar):
    fig, ax = plt.subplots(figsize=(14, 8))
    
    #Sample every N points for readability (otherwise too many bars)
    sample_rate = max(1, len(df) // 50)
    df_sampled = df.iloc[::sample_rate]
    
    time = df_sampled['time'].values
    X = df_sampled['X'].values * 100
    Y = df_sampled['Y'].values * 100
    Z = df_sampled['Z'].values * 100
    
    width = (time[1] - time[0]) * 0.8 if len(time) > 1 else 1.0
    
    ax.bar(time, X, width=width, label='Hydrogen', color='#FF4444', edgecolor='black', linewidth=0.5)
    ax.bar(time, Y, width=width, bottom=X, label='Helium', color='#FFA500', edgecolor='black', linewidth=0.5)
    ax.bar(time, Z, width=width, bottom=X+Y, label='Metals', color='#FFD700', edgecolor='black', linewidth=0.5)
    ax.set_xlabel('Time (Gyr)', fontsize=14, fontweight='bold')
    ax.set_ylabel('Elemental Composition (%)', fontsize=14, fontweight='bold')
    ax.set_title(f'Elemental Composition Evolution: {M_solar} M☉ Red Dwarf', fontsize=16, fontweight='bold')
    ax.set_ylim(0, 100)
    ax.legend(fontsize=12, loc='upper right')
    ax.grid(True, alpha=0.3, axis='y')
    ax.text(time[0], 105, f'Start\nX={df_sampled["X"].iloc[0]:.3f}', 
            ha='center', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    ax.text(time[-1], 105, f'End\nX={df_sampled["X"].iloc[-1]:.3f}', 
            ha='center', fontsize=10, bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
    
    plt.tight_layout()
    plt.savefig(f'composition_evolution_{M_solar}Msun.png', dpi=300, bbox_inches='tight')
    plt.show()

def plot_star_cross_section_ipywidgets(df, M_solar):
    """
    Interactive cross-section using ipywidgets (had to look this library up; Jupyter Notebook seems to display images natively as PNGs which can't be interacted with)
    """
    
    def draw_star(step):
        plt.figure(figsize=(8, 8))
        ax = plt.gca()
        
        X = df['X'].iloc[step]
        Y = df['Y'].iloc[step]
        Z = df['Z'].iloc[step]
        time = df['time'].iloc[step]
        
        r1 = X
        r2 = X + Y
        r3 = X + Y + Z
        
        if Z > 0:
            circle_metals = patches.Circle((0, 0), r3, color='#FFD700', ec='black', linewidth=2, label='Metals (Z)')
            ax.add_patch(circle_metals)
        if Y > 0:
            circle_helium = patches.Circle((0, 0), r2, color='#FFA500', ec='black', linewidth=2, label='Helium (Y)')
            ax.add_patch(circle_helium)
        if X > 0:
            circle_hydrogen = patches.Circle((0, 0), r1, color='#FF4444', ec='black', linewidth=2, label='Hydrogen (X)')
            ax.add_patch(circle_hydrogen)
        
        core = patches.Circle((0, 0), 0.05, color='#333333', ec='black', linewidth=1)
        ax.add_patch(core)
        ax.set_xlim(-1.2, 1.2)
        ax.set_ylim(-1.2, 1.2)
        ax.set_aspect('equal')
        ax.set_title(f'{M_solar} M☉ Red Dwarf Cross-Section\n' + f'Time: {time:.2f} Gyr (Step {step}/{len(df)-1})\n' +
                    f'X={X:.3f}, Y={Y:.3f}, Z={Z:.3f}',fontsize=14, fontweight='bold', pad=20)
        ax.legend(loc='upper right', fontsize=11, framealpha=0.9)
        
        ax.set_xticks([])
        ax.set_yticks([])
        for spine in ax.spines.values():
            spine.set_visible(False)
        
        ax.text(0, -1.4, f'H: {X*100:.1f}%  |  He: {Y*100:.1f}%  |  Metals: {Z*100:.1f}%', ha='center', fontsize=12, 
               bbox=dict(boxstyle='round', facecolor='white', alpha=0.8))
        
        plt.tight_layout()
        plt.show()
    
    #This part makes the interactive slider! Try it out.
    interact(draw_star, 
             step=IntSlider(min=0, max=len(df)-1, step=1, value=0, 
                          description='Timestep:', 
                          style={'description_width': '100px'},
                          layout={'width': '500px'}))

plot_composition_evolution(main_df, initial_mass)
plot_star_cross_section_ipywidgets(main_df, initial_mass)
```
